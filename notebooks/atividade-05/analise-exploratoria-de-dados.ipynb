### HIPÓTESES

Tendo em vista que o data set buscava avaliar o tipo de acidez, a utilização de dados nominais como os tipos diferentes de acidez presentes em um vinho parecem não ser relevantes para atingir o resultado esperado.

De fato, os dados de pH aparentam ser proporcionais aos dados de acidez apresentados no dataset, tendo em vista que existe uma variação entre os resultados de pH ao verificar os dados de acidez associados a ele.

FInalmente, como a qualidade é medida por valores inteiros, essa precisão está correta. Nos valores de pH uma precisão de no máximo duas casas decimais é considerada suficiente para uma boa avaliação dos dados. Nos dados restantes, uma precisão de uma ou duas casas decimais é considerada satisfatória.

### INSIGHTS

Existem várias correlações inerentes aos dados apresentados, além dos dados de acidez já citados, a qualidade é medida a partir de uma quantidade ideal de diferentes compostos químicos analisados no nosso data set. É possível verificar que um teor alcoólico ideal vai render um resultado de qualidade mais satisfatório, bem como a quantidade de enxofre, sulfato, açúcar residual, entre outros. Estes compostos em quantidades adequada terão melhores notas em qualidade.
